# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# File 
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# CSV into Pandas Data Frames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)
```

```python
#check that the import worked for the school CSV
school_data_df.head()
```

```python
#check that the import worked for the student CSV
student_data_df.head(10)
```

```python
#define functions for School Summary section later on
# 1 = True and 0 = False 
def passing_reading(col):
    if col['reading_score'] > 69:
        return 1
    else:
        return 0

def passing_math(col):
    if col['math_score'] > 69:
        return 1
    else:
        return 0
student_data_df['passing_reading'] = student_data_df.apply(passing_reading, axis = 1) 
student_data_df['passing_math'] = student_data_df.apply(passing_math, axis=1)
student_data_df.head()
```

```python
# Merge data into a single dataset
school_data_merged = pd.merge(student_data_df, school_data_df, how="outer", on="school_name")
#check that merged DataFrame worked
school_data_merged.head()
```

```python
#rename the merged DataFrame columns 
school_data_complete = school_data_merged.rename(columns=
                                                 {"Student ID":"Student ID",
                                                  "student_name" :"Student Name",
                                                  "gender":"Gender",
                                                  "grade":"Grade",
                                                  "school_name":"School Name",
                                                  "reading_score":"Reading Score",
                                                  "passing_reading" : "Passing Reading",
                                                  "passing_math": "Passing Math",
                                                  "math_score":"Math Score",
                                                  "School ID":"School ID",
                                                  "type":"School Type",
                                                  "size":"Size",
                                                  "budget":"Budget"                                                 
})
#check to see if it worked
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Variables
total_number_of_schools = school_data_df.school_name.value_counts().sum()
total_number_of_students = student_data_df.school_name.value_counts().sum()
total_budget = school_data_df.budget.sum()
average_math_score = round(student_data_df.math_score.mean(), 3)
average_reading_score = round(student_data_df.reading_score.mean(), 3)
overall_average_score = ((average_math_score + average_reading_score)/2)/100
students_passing_math_score = len(student_data_df.loc[student_data_df['math_score'] > 69])
students_passing_reading_score = len(student_data_df.loc[student_data_df['reading_score'] > 69])
passing_math_score_percentage = students_passing_math_score / total_number_of_students
passing_reading_score_percentage = students_passing_reading_score / total_number_of_students

#Create DataFrame with results
district_results_df = pd.DataFrame({
    'Total Schools':[total_number_of_schools],
    'Total Students':[total_number_of_students],
    'Total Budget':[total_budget],
    'Average Math Score':[average_math_score],
    'Average Reading Score':[average_reading_score],
    '% Passing Math':[passing_math_score_percentage],
    '% Passing Reading':[passing_reading_score_percentage],
    '% Overall Passing Rate':[overall_average_score]
})

#format with dollar sign and percentage
district_results_df["Total Budget"] = district_results_df["Total Budget"].astype(float).map("${:,.2f}".format)
district_results_df["Total Students"] = district_results_df["Total Students"].map("{:,}".format)
district_results_df["% Passing Math"] = district_results_df["% Passing Math"].map("{:.2%}".format)
district_results_df["% Passing Reading"] = district_results_df["% Passing Reading"].map("{:.2%}".format)
district_results_df["% Overall Passing Rate"] = district_results_df["% Overall Passing Rate"].map("{:.2%}".format)
#print to view it
district_results_df
```

```python
#format the "Budget" column
school_summary_df_grouped["Budget"] = school_summary_df_grouped["Budget"].astype(float).map("${:,.2f}".format)
school_summary_df_grouped["Per Student Budget"] = school_summary_df_grouped["Per Student Budget"].astype(float).map("${:,.2f}".format)
school_summary_df_grouped.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Reference merged DF
school_data_merged.head()
```

```python
#New spending per student column
school_data_merged["spending_per_student"] = school_data_merged["budget"] / school_data_merged["size"]
school_data_merged.head()
```

```python
#summary stats of spending_per_student
school_data_merged["spending_per_student"].describe()
```

```python
#Based on min/max, create bins 
bins = [575, 605, 635, 665 ,700]
bin_names = ["575-605","605-635","635-665","665-700"]
```

```python
#Binned DF
school_data_merged["binned_budget"] = pd.cut(school_data_merged["spending_per_student"], bins, labels = bin_names)
school_data_merged.head()
```

```python
#Dictionary for AVGs
school_spending_avgs = {'budget':np.mean, 'reading_score':np.mean, 'math_score': np.mean, 'size': np.mean, 'passing_reading': np.sum, 'passing_math': np.sum}
```

```python
#Grouped DF
school_spending_grouped_df = school_data_merged.groupby(["school_name", "binned_budget"], as_index = False).agg(school_spending_avgs)

#per student column
school_spending_grouped_df["spending_per_student"] = school_spending_grouped_df['budget'] / school_spending_grouped_df['size']

#percent of students that passed math
school_spending_grouped_df["%_passing_math"] = (school_spending_grouped_df['passing_math'] / school_spending_grouped_df['size'])

#percent of students that passed reading
school_spending_grouped_df["%_passing_reading"] = (school_spending_grouped_df["passing_reading"] / school_spending_grouped_df["size"])

#overall passing percentage
school_spending_grouped_df["%_passing_overall"] = (school_spending_grouped_df["%_passing_math"] + school_spending_grouped_df["%_passing_reading"]) / 2

#Copy for later
school_spending_grouped_df_2 = pd.DataFrame.copy(school_spending_grouped_df)

#drop columns 
school_spending_grouped_df.drop(["passing_reading", "passing_math"], axis = 1, inplace = True)

school_spending_grouped_df.head()
```

```python
#Clean up column names 
#rename the merged DataFrame columns 
scores_by_school_spending = school_spending_grouped_df.rename(columns=
                                                 {"school_name":"School Name",
                                                  "binned_budget" :"Budget Range $",
                                                  "budget":"Actual Budget",
                                                  "reading_score":"Average Reading Score",
                                                  "math_score":"Average Math Score",
                                                  "size":"Size",
                                                  "spending_per_student" : "Spending Per Student",
                                                  "%_passing_math": "Passing Math %",
                                                  "%_passing_reading":"Passing Reading %",
                                                  "%_passing_overall":"Overall Passing %"
})

#Format Columns
scores_by_school_spending["Actual Budget"] = scores_by_school_spending["Actual Budget"].astype(float).map("${:,.2f}".format)
scores_by_school_spending["Average Reading Score"] = round(scores_by_school_spending["Average Reading Score"], 2)
scores_by_school_spending["Average Math Score"] = round(scores_by_school_spending["Average Math Score"], 2)
scores_by_school_spending["Size"] = scores_by_school_spending["Size"].map("{:,}".format)
scores_by_school_spending["Passing Math %"] = scores_by_school_spending["Passing Math %"].map("{:.2%}".format)
scores_by_school_spending["Passing Reading %"] = scores_by_school_spending["Passing Reading %"].map("{:.2%}".format)
scores_by_school_spending["Overall Passing %"] = scores_by_school_spending["Overall Passing %"].map("{:.2%}".format)
scores_by_school_spending["Spending Per Student"] = scores_by_school_spending["Spending Per Student"].astype(float).map("${:,.2f}".format)

#View
scores_by_school_spending
```

```python
bins = [400, 2000, 3500, 5000]
bin_names = ["Small (400-2000)", "Medium (2000-3500)", "Large (3500-5000)"]
```

```python
#New column.
school_spending_grouped_df_2["school_size"] = pd.cut(school_spending_grouped_df["size"], bins, labels = bin_names)

school_spending_grouped_df_2.head()
```

```python
#Dictionary for avgs
size_calculation = {'budget': np.sum, 'reading_score':np.mean, 'math_score': np.mean, 'size': np.sum, 'passing_reading': np.sum, 'passing_math': np.sum}
```

```python
#groupby 
school_size_grouped = school_spending_grouped_df_2.groupby(['school_size'], as_index = False).agg(size_calculation)

#percetage of students pass math.
school_size_grouped['%_passing_math'] = school_size_grouped['passing_math'] / school_size_grouped['size']

#percetage of students  pass reading.
school_size_grouped['%_passing_reading'] = school_size_grouped['passing_reading'] / school_size_grouped['size'] 

#overall
school_size_grouped['%_passing_overall'] = (school_size_grouped['%_passing_math'] + school_size_grouped['%_passing_reading']) / 2
#drop 
school_size_grouped.drop(['passing_reading', 'passing_math','budget','size'], axis = 1, inplace = True)

#Return the new dataframe. 
school_size_grouped
```

```python
#Clean up column names 
#rename the merged DataFrame columns 
scores_by_school_size = school_size_grouped.rename(columns=
                                                 {"school_size":"School Size",
                                                  "reading_score" :"Avg Reading Score",
                                                  "math_score":"Avg Math Score",
                                                  "%_passing_math":"Passing Math %",
                                                  "%_passing_reading":"Passing Reading %",
                                                  "%_passing_overall":"Passing Overall %"
})

#Format Columns
scores_by_school_size["Avg Reading Score"] = round(scores_by_school_size["Avg Reading Score"], 2)
scores_by_school_size["Avg Math Score"] = round(scores_by_school_size["Avg Math Score"], 2)
scores_by_school_size["Passing Math %"] = scores_by_school_size["Passing Math %"].map("{:.2%}".format)
scores_by_school_size["Passing Reading %"] = scores_by_school_size["Passing Reading %"].map("{:.2%}".format)
scores_by_school_size["Passing Overall %"] = scores_by_school_size["Passing Overall %"].map("{:.2%}".format)

#View
scores_by_school_size
```
